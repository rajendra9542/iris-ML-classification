```python
# Unnnecessary of ID column. To drop it now
df = df.drop("Id",axis=1)
df.head()
```

Step 4: Separating I/p and O/p

```python
x = df.drop("Species",axis=1)
y=df["Species"]
print(x)  # I/P
print(y)  # O/P
```

Step 5️⃣ Split Train & Test Data

Usually 80% training, 20% testing.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Step 6️⃣ Which Algorithm to Use?

Since this is classification, good algorithms are:

Algorithm	Good?	Why

Logistic Regression	✅	Fast & Simple

Decision Tree	✅	Easy to understand

Random Forest	⭐ Best	High Accuracy

KNN	⭐ Very Famous for Iris	Works well small dataset

SVM	⭐ Excellent	Great for classification

Step 7️⃣ Train Model

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

Step 8️⃣ Predict Test Data

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)
```
